# **III. EJECUTAR PROCESO**

```python
import requests
import ipywidgets as widgets
from IPython.display import display, HTML


def obtener_tasa_cambio_api(moneda_origen, moneda_destino):
  """
  Obtiene la tasa de cambio entre dos monedas utilizando una API de tasas de cambio.

  Args:
  moneda_origen (str): Código de la moneda de origen (ej. 'USD').
  moneda_destino (str): Código de la moneda de destino (ej. 'NIO').

  Returns:
  str: Tasa de cambio formateada o mensaje de error.
  """
  api_key = 'TU_API_KEY_AQUI'
  url = f"https://api.exchangerate-api.com/v4/latest/{moneda_origen}"

  try:
      response = requests.get(url)
      response.raise_for_status()
      data = response.json()

      if moneda_destino in data['rates']:
          tasa_cambio = data['rates'][moneda_destino]
          return f"1 {moneda_origen} ({monedas[moneda_origen]}) = {tasa_cambio} {moneda_destino} ({monedas[moneda_destino]})"
      else:
          return f"No se pudo encontrar la tasa de cambio para {moneda_origen} a {moneda_destino}"

  except requests.RequestException as e:
      return f"Error al obtener la tasa de cambio: {str(e)}"

# Crear opciones para los dropdowns
opciones_monedas = [(f"{codigo} - {nombre}", codigo) for codigo, nombre in sorted(monedas.items())]

# Crear widgets
dropdown_origen = widgets.Dropdown(
  options=opciones_monedas,
  description='Moneda de origen:',
  style={'description_width': 'initial'}
)

dropdown_destino = widgets.Dropdown(
  options=opciones_monedas,
  description='Moneda de destino:',
  style={'description_width': 'initial'}
)

boton_convertir = widgets.Button(description="Convertir")
output = widgets.Output()

def on_button_clicked(b):
  with output:
      output.clear_output()
      moneda_origen = dropdown_origen.value
      moneda_destino = dropdown_destino.value
      resultado = obtener_tasa_cambio_api(moneda_origen, moneda_destino)
      print(resultado)

boton_convertir.on_click(on_button_clicked)
```

# **IV. CONVERTIR MONEDA**

```python
display(HTML("<h2>Conversor de Monedas</h2>"))
display(dropdown_origen, dropdown_destino, boton_convertir, output)
```
